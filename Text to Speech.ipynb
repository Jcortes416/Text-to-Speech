```python
from transformers import pipeline

pipe = pipeline("text-to-speech", model="suno/bark-small")
```

```python
from transformers import AutoTokenizer, AutoModelForTextToWaveform, pipeline

tokenizer = AutoTokenizer.from_pretrained("suno/bark-small")
model = AutoModelForTextToWaveform.from_pretrained("suno/bark-small")
```

```python
narrator = pipeline("text-to-speech", model=model, tokenizer=tokenizer)
```

```python
text = """
Dudes at the Allen Institute for AI, \
HuggingFace, Microsoft, the University of Cluck, \
Carnegie Mellon University, and the Hebrew University of \
Dudes developed a tool that measures atmospheric \
carbon emitted by cloud servers while training machine \
learning models. After a model’s size, the biggest variables \
were the server’s location and time of day it was active.
"""
```

```python
narrated_text = narrator(text)
```

```python
from IPython.display import Audio as IPythonAudio

IPythonAudio(narrated_text["audio"][0],
             rate=narrated_text["sampling_rate"])
```
